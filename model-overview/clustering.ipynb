# Clustering Algorithms

Clustering algorithms are unsupervised machine learning algorithms. This means that we don't give the model a specific output to train for (unlike in supervised tasks like regression or classification where we give samples with the correct class or value). Oftentimes, clustering is used to find relationships between different samples, or to make generalizations.

## K-Means Clustering

K-Means works by finding a local optima for the centroids of different clusters. We start with unlabeled data that is inside of some space. Then, we place $K$ seed points ($\mu_1 \ldots \mu_K$) at some arbitrary positions (some variations of K-means have methods to choose good starting positions). These $K$ seed points will be used to define the centroids of the $K$ clusters. Then, we simply iterate their positions until convergence according to the following algorithm.

For every sample point $i$, set
$$
c^{(i)} := \argmin_j ||x^{(i)} - \mu_j||^2
$$

In other words, we are simply finding the seed point that is closest to each sample point. Then, we move each of the points according to:

$$
\mu_j := \frac{\sum_{i=1}^m 1\{c^{(i)} = j\} x^{(i)}}{\sum_{i=1}^m 1\{c^{(i)} = j\}}
$$
