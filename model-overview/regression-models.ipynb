# All Common Regression Machine Learning Algorithms

This document outlines all the commonly used machine learning algorithms for regression problems where we are trying to produce a best-fit line that gives expected value output given inputs.

## Linear Regression

Linear regression multiples an n-dimensional (input dim) weight vector $\vec{w}$ to the input vector (so multiplying each input value by a weight), and adds on an additional bias. This result is the predicted value. 
$$
\hat{y} = \vec{w} \cdot \vec{x} + b
$$

$$
\hat{y} = w_1x_1 + w_2x_2 + \ldots + w_nx_n + b
$$

It uses the mean-squared error (MSE) cost function, which sums up the square of each difference between the true value and predicted value. 
$$
J = MSE = \frac{1}{m} \sum_{i=1}^{m} (y_i - \hat{y_i})^2
$$
