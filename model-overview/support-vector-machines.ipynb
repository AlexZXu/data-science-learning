# Support Vector Machines

## SVMs for Classification

The focus of support vector machines is to create a line that maximizes the __soft margin__. In other words, we want as much gap between the dividng line and the first classification points (called support vectors) as possible. 

Like regular logistic regression, our decision rule is given by:
$$
\vec{w} \cdot \vec{x} + b \geq 0
$$

However, with SVM, we define a new margin. For any samples that are of class +1, we say that $\vec{w} \cdot {x_+} + b \geq 1$, and for any samples that are of class -1, we say that $\vec{w} \cdot {x_-} + b \leq -1$. Any samples that do not follow this rule are either in the gutter or on the wrong side. We can simplify this by creating a helper variable:

$$
y_i = 
\begin{cases}
    +1 & \text{if } x_+ \\
    -1  & \text{if } x_-
\end{cases}
$$

Then, we have the general equation:

$$
y_i(\vec{w} \cdot x_i + b) \geq 1
$$

We can compute the width of the margin by imagining two support vectors $x_i$ and $x_j$ that are on both sides of the margin (one on the + side of the margin and one on the - side of the margin). The vector difference $x_i - x_j$ then spans the entire margin, but not perpendicularly. One way to find the perpendicular length is to project the vector $x_i - x_j$ onto a unit vector that is orthogonal to the boundary line. The boundary line is $\vec{w} \cdot \vec{x} + b$, so conveniently, $\vec{w}$ is orthogonal to the boundary line. Thus, we can convert $\vec{w}$ into a unit vector using $\frac{\vec{w}}{||w||}$.

Thus, we have:

$$
width = (x_i - x_j) \cdot \frac{\vec{w}}{||w||}
$$

$x_i$ and $x_j$ are on the boundary of the margins, so they satisfy $y_i(\vec{w} \cdot x_i + b) = 1$. With a bit of manipulation, we obtain $w \cdot x_i (+) = 1 - b$ and $w \cdot x_j (-) = -1 - b$. Thus, we have:

$$
\frac{w \cdot x_i - w \cdot x_j}{||w||} \implies
$$
$$
\frac{(1 - b) - (-1 - b)}{||w||} \implies
$$
$$
\frac{2}{||w||}
$$

Thus, it turns out that the total width of the margin is just $\frac{2}{||w||}$. Because the goal of SVM is to maximize this margin, we want to maximize $\frac{2}{||w||}$, which is equivalent to minimizing $||w||$, which (for mathematical convenience), is equivalent to minimizing $\frac{1}{2}||w||^2$


However, we need to balance our desire to maximize the margin with the cost of misclassification. The more negative $y_i(\vec{w} \cdot x_i + b)$ is, the more off a sample is from the correct boundary. Thus, we naturally define the loss of a single sample as:

$$
\max(0, 1 - y_i(\vec{w} \cdot x_i + b))
$$

Note that when a sample is classified properly (and is outside of the margin space), $y_i(\vec{w} \cdot x_i + b) > 1$, so $1 - y_i(\vec{w} \cdot x_i + b)< 0$, and the max function just picks 0 loss.

Thus, the classic optimization problem for SVM is:

$$
L = \frac{1}{2}||w||^2 + C\sum_{i = 1}^{m} \max(0, 1 - y_i(\vec{w} \cdot x_i + b))
$$

$$
\argmin_{\vec{w}, b} L
$$


Note that the $C$ is a constant that tells us how much we want to care about the misclassifications relative to the margin width.

## Kernel Trick

A major problem is that it is often infeasible to fit a single hyperplane to split the sample data on the same input dimension. Thus, a trick we can do is to transform the data onto a higher dimension that we can slice through. A few standard kernels are:

- Linear kernel (basically no kernel): $x_i \cdot x_j$
- Polynomial kernel: $(x_i \cdot x_j + r)^d$
- Radial Basis Function (RBF) Kernel: $exp(\gamma||x_i-x_j||^2)$ ($\gamma$ is a constant we can search through)
- Sigmoid Kernel: $\tanh(\gamma x_i \cdot x_j + r)$

There is complicated math involving Lagrange multipliers (that I won't get into) that show that you can fit the SVM using the Hinge Loss function defined above using just the kernel computations instead of actually transforming the data onto the higher dimension and then performing SVM.

## SVMs for Regression

In regression, a similar idea of margin is used. In this case, we allow a buffer tube of radius $\epsilon$ where we accept errors with no loss, so that we construct a more generalized regression curve. In other words, our prediction is:

$$
\vec{w} \cdot \vec{x} + b
$$

and we say that the loss is 0 if:

$$
|y_i - (\vec{w} \cdot \vec{x} + b)| \leq \epsilon
$$

Using the same hinge loss idea from classification, the loss of a single sample is defined as:

$$
\max(0, |y_i - (\vec{w} \cdot \vec{x} + b)| - \epsilon) \implies \max(0, |y_i - f(x_i)| - \epsilon)
$$

The total loss is similarly defined as:

$$
L = \frac{1}{2}||w||^2 + C\sum_{i = 1}^{m} \max(0, |y_i - f(x_i)| - \epsilon)
$$

Note that in case the first term with $||w||$ has less to directly do with maximizing the margin (although it is still related) and more to do with simple regularization (just L2 regularization with $\lambda = 0.5$).

We can do the same kernel trick on SVR, though its application is pretty complicated.
